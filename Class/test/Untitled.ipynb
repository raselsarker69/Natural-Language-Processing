```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_diagram():
    fig, ax = plt.subplots(figsize=(4, 6))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 15)
    ax.axis("off")
    
    # Boxes with text
    layers = [
        ("Input", 14, "white"),
        ("Attention Block", 11, "#003366"),
        ("Feature Fusion Layer", 8, "#003366"),
        ("Prediction Layer", 5, "#003366"),
        ("Sentiment Decoder", 2, "#008066")
    ]
    
    # Draw input box with extended lines
    ax.plot([2, 8], [13.5, 13.5], color='black', lw=2)
    ax.plot([2, 2], [13.5, 12.5], color='black', lw=2)
    ax.plot([8, 8], [13.5, 12.5], color='black', lw=2)
    ax.text(5, 13.8, "Input", fontsize=12, color="black", ha='center', va='center', fontweight='bold')
    
    for text, y, color in layers[1:]:
        rect = patches.Rectangle((2, y), 6, 2, edgecolor="black", facecolor=color, lw=2)
        ax.add_patch(rect)
        ax.text(5, y + 1, text, fontsize=12, color="white", ha='center', va='center', fontweight='bold')
    
    # Arrows
    arrow_y_positions = [12, 9, 6, 3]
    for y in arrow_y_positions:
        ax.arrow(5, y, 0, -1, head_width=0.4, head_length=0.5, fc="black", ec="black")
    
    plt.show()

draw_diagram()
```
