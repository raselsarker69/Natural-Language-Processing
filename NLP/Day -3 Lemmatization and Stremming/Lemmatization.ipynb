## Lemmatization in Natural Language Processing
- Lemmatization refers to reducing words to their base or dictionary form, known as the lemma, using a vocabulary and morphological analysis. Unlike stemming, lemmatization removes only inflectional endings while ensuring that the root word (lemma) is meaningful.
- Lemmatization reduces words to their base word, which is linguistically correct lemmas. It transforms root word with the use of vocabulary and morphological analysis. 

Lemmatization: Finding the same root 
- Input : (Sings, Sung, Sang)
- Output : Sing

### Difference Between Stemming and Lemmatization
---
**Stemming:**  
স্ট্রিমিং হলো একটি প্রক্রিয়া যেখানে শব্দগুলোর শেষের অংশ কেটে ফেলে তাদের মূল রূপে (root form) রূপান্তরিত করা হয়। তবে স্ট্রিমিংয়ের ক্ষেত্রে যে রূপে শব্দটি রূপান্তরিত হয়, তা সবসময় meaningful শব্দ নাও হতে পারে।  
**Example:**  
- "playing" শব্দটি স্ট্রিমিংয়ের মাধ্যমে "play" এ রূপান্তরিত হতে পারে।  
- তবে "better" শব্দটি স্ট্রিমিংয়ের মাধ্যমে "bett" এ রূপান্তরিত হয়ে অর্থহীন হয়ে যেতে পারে।  

**Lemmatization:**  
অন্যদিকে, লেমাটাইজেশন শব্দগুলোকে তাদের মূল রূপ বা "lemma" তে রূপান্তরিত করে এবং এটি সবসময় একটি meaningful শব্দ হিসেবে গ্রহণযোগ্য হতে হবে। Lemmatization প্রক্রিয়াটি শব্দের ব্যাকরণগত বৈশিষ্ট্য এবং প্রসঙ্গ অনুযায়ী শব্দটিকে রূপান্তর করে, ফলে শব্দটি একটি সঠিক ও অর্থবোধক রূপে থাকে।  
**Example:**  
- "playing" শব্দটি লেমাটাইজেশনের মাধ্যমে "play" এ রূপান্তরিত হয়।  
- "better" শব্দটি লেমাটাইজেশনের মাধ্যমে "good" এ রূপান্তরিত হয়, যা একটি meaningful এবং সঠিক রূপ।  

### Key differences:
- **Stemming:** শব্দের শেষাংশ কেটে ফেলা হয়, যা অনেক সময় অর্থহীন হতে পারে।  
- **Lemmatization:** শব্দকে ব্যাকরণ ও প্রসঙ্গভিত্তিকভাবে রূপান্তরিত করে, যা সবসময় meaningful হয়।

### N.B: Lemmatization অনেক বেশি grammatically কারেক্ট এবং accuracy অনেক high.

### Features of lemmatization and streaming

**Lemmatization:**  
লেমাটাইজেশন অনেক বেশি ব্যাকরণগত (grammatical) নিয়ম অনুসরণ করে। এটি শব্দের প্রসঙ্গ এবং ব্যাকরণ অনুযায়ী কাজ করে এবং যে শব্দটি লেমাটাইজেশনের মাধ্যমে পাওয়া যায়, তা সবসময় একটি অর্থপূর্ণ (meaningful) শব্দ হয়ে থাকে।  

**Stemming:**  
স্ট্রিমিং অনেক বেশি আক্রমণাত্মক (aggressive) পদ্ধতি। এটি সাধারণত কমন শব্দতাত্ত্বিক (morphological) কাঠামোর দিকে বেশি মনোযোগ দেয় না। ফলে যে রূপে শব্দটি রূপান্তরিত হয়, তা অনেক সময় অর্থহীন হতে পারে। স্ট্রিমিং তুলনামূলকভাবে দ্রুত এবং কম জটিল।  

### Main difference:
- **Lemmatization:** 
  - grammar মেনে চলে। 
  - শব্দের রূপ সবসময় অর্থপূর্ণ হয়। 
  - প্রসঙ্গ অনুযায়ী শব্দের সঠিক রূপ নির্ধারণ করে।
  - accuracy high.
  - Lemmatization always slower.
  - Huge amount/Large dataset a কাজ করতে গেলে Lemmatization a problems হয়

- **Stemming:** 
  - অনেক বেশি aggressive and faster। 
  - শব্দের কাঠামো বা অর্থপূর্ণতা সবসময় নিশ্চিত করে না। 
  - কমন morphological কাঠামো যাচাই করে না।

### বড় ডেটাসেট বা বিশাল পরিমাণ ডেটার ক্ষেত্রে লেমাটাইজেশনের সমস্যা

**লেমাটাইজেশন (Lemmatization):**  
লেমাটাইজেশন একটি জটিল এবং ব্যাকরণগতভাবে সঠিক প্রক্রিয়া। এটি প্রতিটি শব্দের ব্যাকরণ, প্রসঙ্গ, এবং অর্থ বিশ্লেষণ করে তাদের মূল রূপে রূপান্তর করে। বড় ডেটাসেট বা বিশাল পরিমাণ ডেটা নিয়ে কাজ করার সময় লেমাটাইজেশনের কিছু সমস্যা দেখা দিতে পারে, যেমন:

#### **সমস্যাগুলো:**
1. **ধীরগতির প্রক্রিয়া:**  
   - লেমাটাইজেশন ব্যাকরণ এবং প্রসঙ্গ অনুযায়ী শব্দ বিশ্লেষণ করে, যা প্রচুর সময় নেয়। বড় ডেটাসেটের ক্ষেত্রে এটি আরও ধীরগতির হয়ে ওঠে।  

2. **অধিক প্রসেসিং পাওয়ার প্রয়োজন:**  
   - লেমাটাইজেশনের জন্য উচ্চতর প্রসেসিং ক্ষমতা এবং মেমোরি প্রয়োজন, বিশেষ করে বিশাল ডেটাসেটের ক্ষেত্রে।  

3. **জটিল অ্যালগরিদম:**  
   - লেমাটাইজেশন প্রক্রিয়া স্টেমিং-এর তুলনায় অনেক বেশি জটিল, যা বড় ডেটাসেটের সাথে কাজ করার সময় কর্মক্ষমতার উপর প্রভাব ফেলে।  

4. **কাস্টমাইজেশনের অভাব:**  
   - বড় ডেটাসেটে থাকা বিভিন্ন ধরনের ডোমেইন-স্পেসিফিক শব্দ লেমাটাইজেশন প্রক্রিয়া সঠিকভাবে হ্যান্ডেল করতে পারে না।  

#### **সমাধান:**
- **স্ট্রিমিং ব্যবহার:**  
  বড় ডেটাসেটের ক্ষেত্রে স্ট্রিমিং ব্যবহার করা যেতে পারে কারণ এটি দ্রুত এবং কম রিসোর্স ব্যবহার করে।
  
- **প্রি-প্রসেসিং:**  
  বড় ডেটাসেটের প্রি-প্রসেসিংয়ের সময় লেমাটাইজেশনের পরিবর্তে স্ট্রিমিং ব্যবহার করে প্রসেসিং সময় এবং রিসোর্সের ব্যবহার কমানো যায়।  

- **বিকল্প প্রযুক্তি ব্যবহার:**  
  স্পেসি (SpaCy) বা অন্যান্য অপ্টিমাইজড টুল ব্যবহার করে লেমাটাইজেশন প্রক্রিয়াটি আরও দ্রুত করা সম্ভব।  

### উপসংহার:
লেমাটাইজেশন বড় ডেটাসেটের ক্ষেত্রে ধীর এবং রিসোর্স-সাশ্রয়ী নয়। এই পরিস্থিতিতে স্ট্রিমিং বা অন্য সহজতর প্রক্রিয়া ব্যবহার করা কার্যকর হতে পারে।

```python
from nltk.stem import WordNetLemmatizer
from textblob import Word
import spacy
from gensim.utils import lemmatize

# Example sentence
sentence = "The striped bats are hanging on their feet for best."

# Tokenize words
words = sentence.split()

# 1. WordNet Lemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
wordnet_lemmas = [wordnet_lemmatizer.lemmatize(word) for word in words]
print("WordNet Lemmatizer:", wordnet_lemmas)

# 2. SpaCy Lemmatizer
nlp = spacy.load("en_core_web_sm")
doc = nlp(sentence)
spacy_lemmas = [token.lemma_ for token in doc]
print("SpaCy Lemmatizer:", spacy_lemmas)

# 3. TextBlob Lemmatizer
textblob_lemmas = [Word(word).lemmatize() for word in words]
print("TextBlob Lemmatizer:", textblob_lemmas)

# 4. Gensim Lemmatizer
gensim_lemmas = [lemmatize(word)[0].decode('utf-8').split('/')[0] for word in words if lemmatize(word)]
print("Gensim Lemmatizer:", gensim_lemmas)

# Note: TreeTagger requires separate installation and configuration.
```

```python
from nltk.stem import WordNetLemmatizer
from textblob import Word
import spacy
from gensim.utils import lemmatize

# Install required packages:
# pip install nltk spacy textblob gensim
# python -m spacy download en_core_web_sm

# Example sentence
sentence = "The striped bats are hanging on their feet for best."

# Tokenize words
words = sentence.split()

# 1. WordNet Lemmatizer
print("---- WordNet Lemmatizer ----")
wordnet_lemmatizer = WordNetLemmatizer()
wordnet_lemmas = [wordnet_lemmatizer.lemmatize(word) for word in words]
print("WordNet Lemmatizer Output:", wordnet_lemmas)

# 2. SpaCy Lemmatizer
print("\n---- SpaCy Lemmatizer ----")
nlp = spacy.load("en_core_web_sm")
doc = nlp(sentence)
spacy_lemmas = [token.lemma_ for token in doc]
print("SpaCy Lemmatizer Output:", spacy_lemmas)

# 3. TextBlob Lemmatizer
print("\n---- TextBlob Lemmatizer ----")
textblob_lemmas = [Word(word).lemmatize() for word in words]
print("TextBlob Lemmatizer Output:", textblob_lemmas)

# 4. Gensim Lemmatizer
print("\n---- Gensim Lemmatizer ----")
gensim_lemmas = []
for word in words:
    try:
        gensim_lemmas.extend([lem.decode("utf-8").split('/')[0] for lem in lemmatize(word)])
    except:
        pass
print("Gensim Lemmatizer Output:", gensim_lemmas)

# TreeTagger requires a separate installation and language model setup
print("\n---- TreeTagger ----")
print("TreeTagger lemmatization requires additional setup and is not shown here.")
```
