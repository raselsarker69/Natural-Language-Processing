```
N-grams
üîπ N-gram ‡¶™‡¶¶‡ßç‡¶ß‡¶§‡¶ø‡¶∞ ‡¶Æ‡¶æ‡¶ß‡ßç‡¶Ø‡¶Æ‡ßá number of combination, number of sequence, number of features ‡¶ú‡ßá‡¶®‡¶æ‡¶∞‡ßá‡¶ü ‡¶ï‡¶∞‡¶§‡ßá ‡¶™‡¶æ‡¶∞‡¶ø‡•§ ‡¶è‡¶ü‡¶ø text data ‡¶ï‡ßá ‡¶Ü‡¶∞‡ßã smoothly analysis ‡¶ï‡¶∞‡¶§‡ßá ‡¶∏‡¶ï‡ßç‡¶∑‡¶Æ ‡¶ï‡¶∞‡ßá‡•§
‡¶Ö‡¶∞‡ßç‡¶•‡¶æ‡ßé ‡¶Ü‡¶Æ‡¶ø ‡¶Ü‡¶Æ‡¶æ‡¶∞ NLP Model ‡¶ï‡ßá ‡¶¨‡¶ø‡¶≠‡¶ø‡¶®‡ßç‡¶® ‡¶Ö‡ßç‡¶Ø‡¶æ‡¶ô‡ßç‡¶ó‡ßá‡¶≤ ‡¶•‡ßá‡¶ï‡ßá different word sequences ‡¶è‡¶∞ ‡¶Æ‡¶æ‡¶ß‡ßç‡¶Ø‡¶Æ‡ßá train ‡¶ï‡¶∞‡¶æ‡¶§‡ßá ‡¶™‡¶æ‡¶∞‡¶§‡ßá‡¶õ‡¶ø‡•§ ‡¶è‡¶á concept ‡¶ü‡¶æ ‡¶ï‡ßá N-grams ‡¶¨‡¶≤‡ßá‡•§

üîπ N-grams ‡¶è‡¶ï‡¶ü‡¶æ sentence ‡¶ï‡ßá ‡¶®‡¶æ‡¶Æ‡ßç‡¶¨‡¶æ‡¶∞ ‡¶Ö‡¶´ N-grams ‡¶è‡¶∞ ‡¶â‡¶™‡¶∞ ‡¶°‡¶ø‡¶™‡ßá‡¶®‡ßç‡¶° ‡¶ï‡¶∞‡ßá Bigram ‡¶π‡¶¨‡ßá ‡¶®‡¶æ‡¶ï‡¶ø Trigrams ‡¶π‡¶¨‡ßá ‡¶®‡¶æ‡¶ï‡¶ø Four-grams ‡¶π‡¶¨‡ßá ‡¶è‡¶ü‡¶æ‡¶∞ ‡¶â‡¶™‡¶∞ depend ‡¶ï‡¶∞‡ßá ‡¶∏‡ßá ‡¶Ö‡¶®‡ßá‡¶ï‡¶ó‡ßÅ‡¶≤‡ßã sequence of words ‡¶ú‡ßá‡¶®‡¶æ‡¶∞‡ßá‡¶ü ‡¶ï‡¶∞‡ßá ‡¶¶‡¶ø‡¶ö‡ßç‡¶õ‡ßá ‡¶Ø‡ßá‡¶ü‡¶æ ‡¶è‡¶®‡¶è‡¶≤‡¶™‡¶ø ‡¶Æ‡¶°‡ßá‡¶≤‡ßá‡¶∞ ‡¶®‡¶æ‡¶Æ‡ßç‡¶¨‡¶æ‡¶∞ of features ‡¶π‡¶ø‡¶∏‡ßá‡¶¨‡ßá ‡¶ï‡¶æ‡¶ú ‡¶ï‡¶∞‡¶¨‡ßá‡•§

Example:
Sentence: "This burger is very tasty"
üîπ 1-Gram: ['This', 'burger', 'is', 'very', 'tasty']
üîπ 2-Gram (Bigram): ['This burger', 'burger is', 'is very', 'very tasty']
üîπ 3-Gram (Trigram): ['This burger is', 'burger is very', 'is very tasty']
üîπ 4-Four-gram: ['This burger is very', 'burger is very tasty']

N-grams Vectorization
üîπ Similar to the count vectorization technique, in the N-Gram method, a document term matrix is generated, and each cell represents the count.
üîπ The columns represent all columns of adjacent words of length n.
üîπ Count vectorization is a special case of N-Gram where n = 1.
üîπ N-grams consider the sequence of n words in the text, where n is (1, 2, 3‚Ä¶). Like 1-gram, 2-gram for token pair. Unlike BoW, it maintains word order.

<div style="background-color: #f9f9fc; color: #333366; border-radius: 12px; margin: 20px auto; padding: 20px; border: 2px solid #ff4c4c; max-width: 1000px; font-family: Arial, sans-serif; line-height: 1.6;">
  <h2 style="text-align: center; color: #333366;">TF-IDF Vectorizer</h2>

---

- TF: ‡¶Ø‡ßá‡¶ï‡ßã‡¶®‡ßã ‡¶°‡¶ï‡ßÅ‡¶Æ‡ßá‡¶®‡ßç‡¶ü‡ßá Total ‡¶ï‡¶Ø‡¶º‡¶ü‡¶æ unique words ‡¶Ü‡¶õ‡ßá ‡¶è‡¶¨‡¶Ç total words ‡¶ï‡¶Ø‡¶º‡¶ü‡¶æ ‡¶Ü‡¶õ‡ßá ‡¶§‡¶æ‡¶∞ ‡¶Ö‡¶®‡ßÅ‡¶™‡¶æ‡¶§‡¶ï‡ßá TF ‡¶¨‡¶≤‡ßá

## 1. Term Frequency (TF):

TF: ‡¶è‡¶ï‡¶ü‡¶ø ‡¶°‡¶ï‡ßÅ‡¶Æ‡ßá‡¶®‡ßç‡¶ü‡ßá ‡¶è‡¶ï‡¶ü‡¶ø Unique word ‡¶ï‡¶§‡¶¨‡¶æ‡¶∞ ‡¶è‡¶∏‡ßá‡¶õ‡ßá, ‡¶§‡¶æ‡¶∞ ‡¶™‡¶∞‡¶ø‡¶Æ‡¶æ‡¶™‡•§. The formula is:

$$
TF = \frac{\text{Number of times the word appears in the document}}{\text{Total number of words in the document}}
$$

- Document-01: "I love Natural Language Processing"
- Document-02: "I love computer vision"
  
- ‡¶Æ‡ßã‡¶ü ‡¶∂‡¶¨‡ßç‡¶¶ ‡¶∏‡¶Ç‡¶ñ‡ßç‡¶Ø‡¶æ: 5‡¶ü‡¶ø ['I', 'love', 'Natural', 'Language', 'Processing']
# 
Document-01 ‡¶è‡¶∞ ‡¶™‡ßç‡¶∞‡¶§‡¶ø‡¶ü‡¶ø ‡¶∂‡¶¨‡ßç‡¶¶‡ßá‡¶∞ TF = 0.2, ‡¶ï‡¶æ‡¶∞‡¶£ ‡¶∏‡¶¨ ‡¶∂‡¶¨‡ßç‡¶¶ ‡ßß ‡¶¨‡¶æ‡¶∞ ‡¶è‡¶∏‡ßá‡¶õ‡ßá ‡¶è‡¶¨‡¶Ç ‡¶Æ‡ßã‡¶ü ‡ß´‡¶ü‡¶ø ‡¶∂‡¶¨‡ßç‡¶¶ ‡¶∞‡¶Ø‡¶º‡ßá‡¶õ‡ßá‡•§

---

## 2. Inverse Document Frequency (IDF):

- IDF: ‡¶∏‡ßá‡¶á ‡¶∂‡¶¨‡ßç‡¶¶‡¶ü‡¶ø ‡¶ï‡¶§‡¶ó‡ßÅ‡¶≤‡ßã ‡¶°‡¶ï‡ßÅ‡¶Æ‡ßá‡¶®‡ßç‡¶ü‡ßá ‡¶™‡¶æ‡¶ì‡¶Ø‡¶º‡¶æ ‡¶Ø‡¶æ‡¶Ø‡¶º, ‡¶§‡¶æ‡¶∞ ‡¶™‡¶∞‡¶ø‡¶Æ‡¶æ‡¶™‡•§. The formula is:
- Total Documents:2

$$
IDF = \log\left(\frac{\text{Total number of documents}}{\text{Number of documents containing the word}}\right)
$$

# 

- Common words (I, love) ‡¶¶‡ßÅ‡¶á ‡¶°‡¶ï‡ßÅ‡¶Æ‡ßá‡¶®‡ßç‡¶ü‡ßá‡¶á ‡¶Ü‡¶õ‡ßá, ‡¶§‡¶æ‡¶á ‡¶§‡¶æ‡¶¶‡ßá‡¶∞ IDF = 0.0‡•§
- Unique words (Natural, Language, Processing) ‡¶ï‡ßá‡¶¨‡¶≤ Document-01 ‡¶è ‡¶Ü‡¶õ‡ßá, ‡¶§‡¶æ‡¶á ‡¶§‡¶æ‡¶¶‡ßá‡¶∞ IDF = 0.693‡•§
---

## Final Formula:
‡¶Ö‡¶∞‡ßç‡¶•‡¶æ‡ßé ‡¶è‡¶ï‡¶ü‡¶æ ‡¶°‡¶ï‡ßÅ‡¶Æ‡ßá‡¶®‡ßç‡¶ü‡ßá TF ‡¶è‡¶¨‡¶Ç IDF ‡¶è‡¶∞ ‡¶ó‡ßÅ‡¶£‡¶´‡¶≤ ‡¶ï‡ßá TF-IDF ‡¶¨‡¶≤‡ßá‡•§
$$
TF - IDF = TF \times IDF
$$

# 
- Common words (I, love) ‡¶è‡¶∞ TF-IDF = 0.0, ‡¶ï‡¶æ‡¶∞‡¶£ ‡¶§‡¶æ‡¶∞‡¶æ ‡¶¶‡ßÅ‡¶á ‡¶°‡¶ï‡ßÅ‡¶Æ‡ßá‡¶®‡ßç‡¶ü‡ßá‡¶á ‡¶∞‡¶Ø‡¶º‡ßá‡¶õ‡ßá‡•§
- Unique words (Natural, Language, Processing) ‡¶è‡¶∞ TF-IDF = 0.1386, ‡¶ï‡¶æ‡¶∞‡¶£ ‡¶§‡¶æ‡¶∞‡¶æ ‡¶∂‡ßÅ‡¶ß‡ßÅ‡¶Æ‡¶æ‡¶§‡ßç‡¶∞ Document-01 ‡¶è ‡¶∞‡¶Ø‡¶º‡ßá‡¶õ‡ßá‡•§
---

## **Why is it important?**
- It highlights important words in a document while reducing the impact of commonly used words.  
- Helps in improving the performance of text-based machine learning models.
