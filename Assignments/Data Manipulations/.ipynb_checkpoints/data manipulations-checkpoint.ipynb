### Removing Rows/Columns

Removing rows and columns using pandas

```python
# import libraries
import pandas as pd

# data
data = {'id':[1,2,3,None,4,None,None,None,None, 10], 'score':[10,20,30,None,None,None,None,None,None, 90]}
print('\nData')
print(data)

# dataframe
df = pd.DataFrame(data)
print('\nData Frame')
print(df)

# drop row
df_drop_rows = df.dropna(thresh=int(0.5*df.shape[1]), axis=0)

# drop column
df_drop_columns = df.dropna(thresh=int(0.5*df.shape[0]), axis=1)

print('\nDrop Null Value From Row')
print(df_drop_rows)
print('\nDrop Null Value From Row')
print(df_drop_columns)
```

### Imputation
Filling missing value with pandas

```python
# import libraries
import pandas as pd

# data
data = {'id':[1,2,3,None,4,None,None,None,None, 10], 'score':[10,20,30,None,None,None,None,None,None, 90]}
print('\nData:')
print(data)

# dataframe
df = pd.DataFrame(data)
print('\nData Frame:')
print(df)

# filling missing value using constant
df_const = df.fillna(0)
print('\nFilling Missing Values Using Constant:')
print(df_const)

# filling missing value using mean
df_mean = df.fillna(df.mean())
print('\nFilling Missing Values Using Mean:')
print(df_mean)

# filling missing value using median
df_median = df.fillna(df.median())
print('\nFilling Missing Values Using Median:')
print(df_median)

# filling missing value using mod
df_mode = df.fillna(df.mode().iloc[0])
print('\nFilling Missing Values Using Mod:')
print(df_mode)
```

Filling missing value with polars

```python
# import libraries
import polars as pl

# data
data = {'id':[1,2,3,None,4,None,None,None,None, 10], 'score':[10,20,30,None,None,None,None,None,None, 90]}
print('\nData:')
print(data)

# dataframe
df = pl.DataFrame(data)
print('\nData Frame:')
print(df)

# filling missing value using constant
df_const = df.fill_null(0)
print('\nFilling Missing Values Using Constant:')
print(df_const)

# filling missing value using mean
df_mean = df.select([pl.col(col).fill_null(pl.col(col).mean()) for col in df.columns])
print('\nFilling Missing Values Using Mean:')
print(df_mean)

# filling missing value using constant
df_median = df.select([pl.col(col).fill_null(pl.col(col).median()) for col in df.columns])
print('\nFilling Missing Values Using Median:')
print(df_median)

# filling missing value using mod
df_mode = df.select([pl.col(col).fill_null(pl.col(col).mode().first()) for col in df.columns])
print('\nFilling Missing Values Using Mod:')
print(df_mode)
```

### Propagation
Forward and backward propagation using pandas

```python
# import libraries
import pandas as pd

# data
data = {'id':[1,2,3,None,4,None,None,None,None, 10], 'score':[10,20,30,None,None,None,None,None,None, 90]}
print('\nData:')
print(data)

# dataframe
df = pd.DataFrame(data)
print('\nData Frame:')
print(df)

# filling missing value using forward propagation
df_forward = df.fillna(method='ffill')
print('\nFilling Missing Values Using forward propagation:')
print(df_forward)

# filling missing value using backward propagation
df_backward = df.fillna(method='bfill')
print('\nFilling Missing Values Using backward propagation:')
print(df_backward)
```

Forward and backward propagation using polars

```python
import polars as pl

# Data
data = {'id': [1, 2, 3, None, 4, None, None, None, None, 10], 'score': [10, 20, 30, None, None, None, None, None, None, 90]}
print('\nData:')
print(data)

# Creating a Polars DataFrame
df = pl.DataFrame(data)
print('\nData Frame:')
print(df)

# Filling missing values using forward propagation (ffill)
df_forward = df.select([pl.col(col).fill_null(strategy='forward') for col in df.columns])
print('\nFilling Missing Values Using Forward Propagation:')
print(df_forward)

# Filling missing values using backward propagation (bfill)
df_backward = df.select([pl.col(col).fill_null(strategy='backward') for col in df.columns])
print('\nFilling Missing Values Using Backward Propagation:')
print(df_backward)
```

### Interpolation
Linear and polynomial interpolation using pandas

```python
# import libraries
import pandas as pd

# data
data = {'id':[1,2,3,None,4,None,None,None,None, 10], 'score':[10,20,30,None,None,None,None,None,None, 90]}
print('\nData:')
print(data)

# dataframe
df = pd.DataFrame(data)
print('\nData Frame:')
print(df)

# filling missing value using linear interpolation
df_linear = df.interpolate(method='linear')
print('\nFilling Missing Values Using linear interpolation:')
print(df_linear)

# filling missing value using polynomial interpolation
df_polynomial = df.interpolate(method='polynomial', order=2)
print('\nFilling Missing Values Using polynomial interpolation:')
print(df_polynomial)
```

## Assignment from Propagation: ffill with avg of recent 3 previous values. ffill with avg of all previous valus

```python
import pandas as pd

# Create DataFrame
data = {'id': [1, 2, 3, 4, 5, None, 7, 8, None, 10], 
        'score': [10, 20, 20, 40, 50, 50, 60, None, 90, 100]}
df = pd.DataFrame(data)

# Iterate through each column in the DataFrame
for col in df.columns:
    nan_indices = df[df[col].isna()].index  # Get indices where NaN values exist

    for idx in nan_indices:
        # Ensure there are at least 3 previous values before NaN
        if idx >= 3:
            prev_values = df.loc[idx - 3:idx - 1, col]  # Get previous 3 values
            mean_value = prev_values.mean()  # Calculate mean
            df.loc[idx, col] = mean_value  # Replace NaN with mean
        else:
            print(f"Not enough previous values to compute mean for index {idx} in column '{col}'.")

# Print modified DataFrame
print(df)
```
